```
from scipy.io import loadmat
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = loadmat('X_data')
X = data['X_data']
X_dataset = pd.DataFrame(data=X)
```

```
data = loadmat('Y_data')
Y = data['Y_data']
Y_dataset = pd.DataFrame(data=Y)
```

```
X_dataset.head()
```

```
x = X_dataset.iloc[800]
y = Y_dataset.iloc[800]
```

```
plt.plot(y, color = 'r',label = 'True spectra')
plt.plot(x, color = 'b',label = 'Predicted spectra')
plt.legend(loc = 'best')
plt.show()
```
